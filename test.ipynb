```python
import numpy as np
import plotly.express as px
from scipy.interpolate import lagrange
import sympy as sp

x_static = [0, 3]
y_static = [0, 3]


def update_plot(x_third, y_third):
    x_points = x_static + [x_third]
    y_points = y_static + [y_third]

    interp_func = lagrange(x_points, y_points)

    x_values = np.linspace(0, 10, 1000)
    y_values = interp_func(x_values)

    return x_values, y_values, interp_func


def tangent(x, x_values, f):
    y = f(x)
    diff = sp.diff(f(sp.Symbol("x")), sp.Symbol("x"))
    f_prime = sp.lambdify(sp.Symbol("x"), diff)
    slope = f_prime(x)

    return slope * (x_values - x) + y


x_values, y_values, f = update_plot(10, 20)
x_tangent = 5

tangent_line = tangent(x_tangent, x_values, f)

fig = px.line(
    x=x_values,
    y=y_values,
    labels={"x": "x", "y": "f(x)"},
    title="Interpolated Function",
)
fig.add_scatter(x=x_values, y=tangent_line, mode="lines", name="Tangent at x=5")

fig.show()
```
